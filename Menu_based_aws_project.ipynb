```python
#tkinter window dimension
root = tk.Tk()
root.title('Shafin Kausar')
root.geometry("1500x800")
root.resizable(True, True)
```

```python
# menu functions : 

def CreateFolder():
    ch = input("File name : ")
    if not (os.path.exists("./"+ch+"/")):
        try:
            os.makedirs('./'+ch+'/')
            print("Folder created successfully!")
        except:
            print("Failed to create Folder.")
    else:
        print("This file already exists, please choose another one!!")
def currentDirectory():
    cwd = os.getcwd()
    value=cwd
    display.config(text=value)

def showAllfolder():
    # global folders_list
    folders_list = os.listdir(os.curdir)
    value = folders_list
    display.config(text=value)

def instanceLaunch():
    instancename = askstring("Name","Enter Instance Name ")
    
    response = messagebox.askquestion("custom Question","Are you sure want to launch ? ",icon='question',type="yesno")
    print("Ec2 instance launched : ",response)
    if response =="yes":
        myec2 = bt.client("ec2")
        response = myec2.run_instances(ImageId = "ami-0ded8326293d3201b",
            InstanceType = "t2.micro",
            MaxCount=1,
            MinCount=1,
            TagSpecifications=[
                    {
                        'ResourceType': 'instance',
                        'Tags': [
                            {
                                'Key': 'Name',
                                'Value': instancename # mypythoOs is the InstanceName
        },]},])
        showinfo('Launched',f"instance {instancename} Launched")
        value = "instance Launched"
        display.config(text=value)
    else:
        print("no response")
        display.config(text="cancelled")

def s3Launch():
    S3 = boto3.client('s3')

    response = S3.create_bucket( 
        ACL = 'private',
        Bucket= 'myteam122',
         CreateBucketConfiguration={
              "LocationConstraint":"ap-south-1"
         }
    )
    value = "S3 Launched"
    display.config(text=value)
    
```

```python

# Creating Menubar
menubar = tk.Menu(root)

# Adding File Menu and commands
file = tk.Menu(menubar, tearoff = 0, font=("Arial", 15))
menubar.add_cascade(label ='File', menu = file)
file.add_command(label ='New File', command = CreateFolder)
file.add_command(label ='Current folder', command = currentDirectory)
file.add_command(label ='show Folders', command = showAllfolder)
#file.add_command(label ='Save', command = None)
file.add_separator()
file.add_command(label ='Exit',background="#FF0080", command = root.destroy)

# Adding Edit Menu and commands
aws = tk.Menu(menubar, tearoff = 0, font=("Arial", 15))
menubar.add_cascade(label ='AWS', menu = aws)
aws.add_command(label ='Instance Launch', command = instanceLaunch)
aws.add_command(label ='S3 Launch', command = s3Launch)


# Adding Help Menu
help_ = tk.Menu(menubar, tearoff = 0, font=("Arial", 15))
menubar.add_cascade(label ='Chill', menu = help_)

help_.add_separator()

```
